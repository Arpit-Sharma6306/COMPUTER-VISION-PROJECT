```python
import cv2
import matplotlib.pyplot as plt

# 1. Image ko load karein
img1 = cv2.imread(r"D:\sachin.jpg", cv2.IMREAD_COLOR)

if img1 is None:
    print("Error: Image load nahi hui! Path sahi se check karein.")
else:
    # 2. OpenCV ka Face Detection Model load karein
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # 3. Image ko Grayscale mein badlein (Detection ke liye zaroori hai)
    gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

    # 4. Face detect karein (Pads ko ignore karne ke liye strict parameters)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=8, minSize=(50, 50))

    # 5. Agar face mil jata hai, toh automatic ussi face ke coordinates par sab draw karein
    for (x, y, w, h) in faces:
        
        # A) Text "Sachin" - Face ke thoda sa upar (y - 20) lagaya hai taaki chehra chhupey nahi
        cv2.putText(img1, "Sachin", (x, y - 20), cv2.FONT_HERSHEY_COMPLEX, 1.2, (255, 255, 0), 2)

        # B) Pehla Rectangle - Khaali Box (White Border) jo face ko cover karega
        cv2.rectangle(img1, (x, y), (x + w, y + h), (255, 255, 255), 3)

        # C) Doosra Rectangle - Poora Bhara Hua Box (Orange/Cyan Solid) 
        # NOTE: -1 ki wajah se yeh face ko poora cover kar dega. 
        # Agar aapko face ke andar ek aur chota border chahiye toh -1 ko badal kar 2 kar dena.
        cv2.rectangle(img1, (x + 15, y + 15), (x + w - 15, y + h - 15), (0, 200, 250), -1)

    # 6. Modified image ko computer mein save karein
    cv2.imwrite("new_sachin_auto.jpg", img1)
    print("Image 'new_sachin_auto.jpg' ke naam se save ho gayi hai!")

    # 7. BGR se RGB badlein aur Jupyter mein display karein
    img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8, 6))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import os

# Video ki settings (Width=640, Height=480, FPS=30)
width, height = 640, 480
fps = 30
duration_sec = 5
total_frames = fps * duration_sec  # Total 150 frames banenge

# Video save karne ka path (F drive)
video_path = r"F:\AI.mp4"

# Agar aapke computer mein F drive nahi hai, toh yeh use D drive mein save karega
if not os.path.exists("F:\\"):
    video_path = r"D:\AI.mp4"  # Backup path

# VideoWriter object jo video file create karega
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(video_path, fourcc, fps, (width, height))

# Loop chala kar video ke frames (images) design karein
for i in range(total_frames):
    # Ek black background image banayein
    frame = np.zeros((height, width, 3), dtype=np.uint8)
    
    # Ek ghumta hua circle (ball) draw karein
    x = int(100 + (i * 3) % (width - 200))
    y = int(240 + 100 * np.sin(i * 0.1))
    
    # Frame par text aur circle draw karein
    cv2.putText(frame, f"Frame: {i}", (20, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.circle(frame, (x, y), 30, (0, 255, 0), -1) # Green color ka ball
    
    # Frame ko video file mein write (add) karein
    out.write(frame)

# Video file ko close karein
out.release()

# FIXED LINE: Yahan se color aur text hata diya hai
print(f"Success! Aapki video yahan save ho gayi hai: {video_path}")
```

```python
import cv2

# 1. Haar Cascade Ka Model Load Karein
# Ensure karein ki "haarcascade_frontalface_default.xml" aapki file ke saath same folder mein ho, 
# ya phir OpenCV ka default path use karein jaise niche diya hai.
face_model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# 2. Webcam Start Karein (0 matlab default primary camera)
video_capture = cv2.VideoCapture(0)

print("Face detection shuru ho gaya hai... Band karne ke liye 'q' dabayein.")

while True:
    # Webcam se frame-by-frame read karein
    ret, frame = video_capture.read()
    if not ret:
        print("Webcam se frame nahi mil raha hai.")
        break

    # OpenCV ke liye image ko Grayscale (black & white) mein convert karna zaroori hota hai
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Faces detect karein
    faces = face_model.detectMultiScale(
        gray_frame,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    # Har detected face ke charo taraf ek rectangle (box) banayein
    for (x, y, w, h) in faces:
        # cv2.rectangle(image, start_point, end_point, color_rgb, thickness)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        
        # Optionally: Aap "Face" text bhi likh sakte hain box ke upar
        cv2.putText(frame, "Face", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Result ko screen par dikhayein
    cv2.imshow('Live Face Detection', frame)

    # Agar 'q' key press hogi, to loop band ho jayega
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Sab khatam hone ke baad camera aur windows ko release karein
video_capture.release()
cv2.destroyAllWindows()
```

```python
import cv2
import tkinter as tk
from tkinter import filedialog, messagebox
import os

class FaceDetectionApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Advanced Face Detection System")
        self.root.geometry("500x450")
        self.root.configure(bg="#2c3e50")

        self.is_video_running = False
        self.video_capture = None

        # --- MODEL INITIALIZATION WITH ERROR HANDLING ---
        # Agar default path kaam na kare to hum error check lagate hain
        self.xml_path = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
        if not os.path.exists(self.xml_path):
            # Backup tarika agar OpenCV default path na mile
            messagebox.showwarning("Warning", "Haar Cascade XML file auto-detect nahi hui. Default settings use ki ja rahi hain.")
        
        self.face_model = cv2.CascadeClassifier(self.xml_path)

        self.login_frame = tk.Frame(self.root, bg="#2c3e50")
        self.main_frame = tk.Frame(self.root, bg="#2c3e50")

        self.show_login_page()

    # ================= PAGE 1: LOGIN PAGE =================
    def show_login_page(self):
        self.main_frame.pack_forget()
        self.login_frame.pack(fill="both", expand=True)

        heading = tk.Label(
            self.login_frame, text="FACE DETECTION", 
            font=("Helvetica", 24, "bold"), fg="#e74c3c", bg="#2c3e50", pady=30
        )
        heading.pack()

        lbl_pass = tk.Label(
            self.login_frame, text="Enter Password:", 
            font=("Arial", 12), fg="#ecf0f1", bg="#2c3e50"
        )
        lbl_pass.pack(pady=10)

        self.entry_pass = tk.Entry(self.login_frame, font=("Arial", 12), show="*", width=25, justify="center")
        self.entry_pass.pack(pady=5)
        self.entry_pass.focus() 

        btn_login = tk.Button(
            self.login_frame, text="LOGIN", font=("Arial", 11, "bold"),
            bg="#2ecc71", fg="white", width=15, height=1, command=self.check_login
        )
        btn_login.pack(pady=20)

        lbl_hint = tk.Label(self.login_frame, text="(Hint: password = admin)", font=("Arial", 9, "italic"), fg="#7f8c8d", bg="#2c3e50")
        lbl_hint.pack(side="bottom", pady=10)

    def check_login(self):
        if self.entry_pass.get() == "admin":
            messagebox.showinfo("Success", "Login Kamyab Raha!")
            self.show_main_page() 
        else:
            messagebox.showerror("Error", "Galat Password! Fir se koshish karein.")
            self.entry_pass.delete(0, tk.END) 

    # ================= PAGE 2: MAIN DASHBOARD =================
    def show_main_page(self):
        self.login_frame.pack_forget() 
        self.main_frame.pack(fill="both", expand=True)

        heading = tk.Label(
            self.main_frame, text="Face Detection Dashboard", 
            font=("Helvetica", 18, "bold"), fg="#ecf0f1", bg="#2c3e50", pady=20
        )
        heading.pack()

        # --- IMAGE MODE ---
        lbl_sec1 = tk.Label(self.main_frame, text="--- Image Mode ---", font=("Arial", 11, "bold"), fg="#3498db", bg="#2c3e50")
        lbl_sec1.pack(pady=5)

        btn_browse = tk.Button(
            self.main_frame, text="Use Image (Select Photo)", font=("Arial", 10, "bold"),
            bg="#34495e", fg="white", width=30, command=self.browse_image
        )
        btn_browse.pack(pady=5)

        # --- LIVE VIDEO MODE ---
        lbl_sec2 = tk.Label(self.main_frame, text="--- Live Video Mode ---", font=("Arial", 11, "bold"), fg="#2ecc71", bg="#2c3e50")
        lbl_sec2.pack(pady=15)

        self.btn_start_vdo = tk.Button(
            self.main_frame, text="START VIDEO", font=("Arial", 11, "bold"),
            bg="#2ecc71", fg="white", width=20, height=2, command=self.start_video
        )
        self.btn_start_vdo.pack(pady=5)

        self.btn_stop_vdo = tk.Button(
            self.main_frame, text="STOP VIDEO", font=("Arial", 11, "bold"),
            bg="#e74c3c", fg="white", width=20, height=2, command=self.stop_video, state="disabled"
        )
        self.btn_stop_vdo.pack(pady=5)

        btn_logout = tk.Button(
            self.main_frame, text="Logout", font=("Arial", 9),
            bg="#95a5a6", fg="black", command=self.logout_and_stop
        )
        btn_logout.pack(side="bottom", pady=15)

    # ================= CORE DETECTION FUNCTIONS =================
    def browse_image(self):
        file_path = filedialog.askopenfilename(
            title="Photo Select Karein",
            filetypes=[("Image Files", "*.jpg *.jpeg *.png *.webp")]
        )
        if file_path:
            img = cv2.imread(file_path)
            if img is None:
                messagebox.showerror("Error", "Image load nahi ho payi! File format check karein.")
                return
            
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            # Parametres thode easy kiye taaki face miss na ho
            faces = self.face_model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3, minSize=(30, 30))
            
            if len(faces) == 0:
                messagebox.showinfo("Result", "Photo me koi face nahi mila! Dusri photo try karein.")
            
            for (x, y, w, h) in faces:
                cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 2)
            
            cv2.imshow("Photo Face Detection - Press ANY KEY to Close", img)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

    def start_video(self):
        if not self.is_video_running:
            self.video_capture = cv2.VideoCapture(0)
            if not self.video_capture.isOpened():
                messagebox.showerror("Error", "Webcam chalu nahi ho raha. Check karein koi aur app camera to use nahi kar raha!")
                return

            self.is_video_running = True
            self.btn_start_vdo.config(state="disabled") 
            self.btn_stop_vdo.config(state="normal")   
            
            # Alag se window pop-up call karne ke liye update loop
            self.update_video_frame()

    def update_video_frame(self):
        if self.is_video_running and self.video_capture.isOpened():
            isFrame, frame = self.video_capture.read()
            if isFrame:
                gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                
                # Live video ke liye optimized face parameters
                faces = self.face_model.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
                
                for (x, y, w, h) in faces:
                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
                    cv2.putText(frame, "Live Face", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)
                
                cv2.imshow("Live Video Feed", frame)
                
                # Agar OpenCV window par 'q' dabayein ya window close karein
                if cv2.waitKey(1) == ord('q'):
                    self.stop_video()
                    return
                
                # Loop ko smooth chalane ke liye 15ms ka gap
                self.root.after(15, self.update_video_frame)
            else:
                self.stop_video()

    def stop_video(self):
        if self.is_video_running:
            self.is_video_running = False
            if self.video_capture:
                self.video_capture.release()
            cv2.destroyAllWindows()
            
            self.btn_start_vdo.config(state="normal")  
            self.btn_stop_vdo.config(state="disabled") 
            messagebox.showinfo("Webcam Status", "Live Video band kar di gayi hai.")

    def logout_and_stop(self):
        self.stop_video()
        self.show_login_page()

if __name__ == "__main__":
    root = tk.Tk()
    app = FaceDetectionApp(root)
    
    # Isse window direct samne khulegi background me nahi chupegi
    root.lift()
    root.attributes('-topmost', True)
    root.after_idle(root.attributes, '-topmost', False)
    
    root.mainloop()
```

```python
import cv2
import numpy as np
import os
from datetime import datetime

# 1. Face dhoondne wala model load karein
face_model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# 2. Check karna ki Step 2 wali training files bani hain ya nahi
if not os.path.exists("trained_faces.npy") or not os.path.exists("trained_ids.npy"):
    print("---------------------------------------------------------------------------")
    print("Error: Pehle naye Step 2 wale code ko run karke faces ko train karein!")
    print("USKE BAAD hi yeh attendance wala code chalega.")
    print("---------------------------------------------------------------------------")
else:
    # Training data load karna
    trained_faces = np.load("trained_faces.npy")
    trained_ids = np.load("trained_ids.npy")

    # Attendance excel/csv file taiyar karna
    if not os.path.exists("attendance.csv"):
        with open("attendance.csv", "w") as f:
            f.write("Name,Date,Time\n")

    # ID ke sath apna naam yahan set karein
    names_dict = {1: "Bhai (Aap)"} 

    def make_entry(name):
        now = datetime.now()
        date_str = now.strftime("%Y-%m-%d")
        time_str = now.strftime("%H:%M:%S")
        
        # Check karna ki aaj attendance pehle se to nahi lagi
        already_marked = False
        with open("attendance.csv", "r") as f:
            lines = f.readlines()
            for line in lines:
                if name in line and date_str in line:
                    already_marked = True
                    break
                    
        if not already_marked:
            with open("attendance.csv", "a") as f:
                f.write(f"{name},{date_str},{time_str}\n")
            print(f"Attendance Recorded Successfully: {name} at {time_str}")

    # Live Webcam on karna
    vdo = cv2.VideoCapture(0)
    print("Attendance Camera ON hai... Band karne ke liye keyboard par 'q' dabayein.")

    while True:
        isFrame, frame = vdo.read()
        if not isFrame:
            break
            
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_model.detectMultiScale(gray, 1.2, 5)
        
        for (x, y, w, h) in faces:
            # Live face ko crop aur resize karna (100x100 size jisse database se match ho sake)
            live_face = gray[y:y+h, x:x+w]
            live_face = cv2.resize(live_face, (100, 100)).flatten()
            
            # Mathematical distance se database ke faces se match nikalna
            distances = np.linalg.norm(trained_faces - live_face, axis=1)
            min_index = np.argmin(distances)
            min_distance = distances[min_index]
            
            # Agar distance 4000 se kam hai to chehra pehchan liya gaya hai
            if min_distance < 4000:  
                user_id = trained_ids[min_index]
                name = names_dict.get(user_id, "Unknown")
                color = (0, 255, 0) # Green Box
                make_entry(name) # Attendance file me naam likhna
            else:
                name = "Unknown Person"
                color = (0, 0, 255) # Red Box
                
            cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
            cv2.putText(frame, f"{name}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
            
        cv2.imshow("Smart Attendance System", frame)
        
        if cv2.waitKey(1) == ord('q'):
            break

    vdo.release()
    cv2.destroyAllWindows()
    print("Camera band ho gaya.")
```

```python
import cv2
import numpy as np
import os
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display, clear_output

# Zaroori folders aur files banana
if not os.path.exists("dataset"):
    os.makedirs("dataset")
if not os.path.exists("attendance.csv"):
    with open("attendance.csv", "w") as f:
        f.write("Name,Date,Time\n")

face_model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
names_dict = {1: "Bhai (Aap)"}
output_area = widgets.Output()

# ================= BACKEND FUNCTIONS =================

def collect_photos(b):
    with output_area:
        clear_output()
        print("Camera shuru ho raha hai... Apne chehre ki 30 photos dene ke liye screen par dekhein.")
        vdo = cv2.VideoCapture(0)
        count = 0
        
        while True:
            isFrame, frame = vdo.read()
            if not isFrame:
                break
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_model.detectMultiScale(gray, 1.3, 5)
            
            for (x, y, w, h) in faces:
                count += 1
                cv2.imwrite(f"dataset/User.1.{count}.jpg", gray[y:y+h, x:x+w])
                cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
                
            cv2.imshow("Photoshoot Mode - Keep Looking", frame)
            if cv2.waitKey(100) == ord('q') or count >= 30:
                break
                
        vdo.release()
        cv2.destroyAllWindows()
        print(f"Kamyabi! 30 Photos successfully save ho gayi hain! Ab Step 2 dabayein.")

def train_model(b):
    with output_area:
        clear_output()
        path = "dataset"
        image_paths = [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.jpg')]
        
        if len(image_paths) == 0:
            print("Error: Dataset khali hai! Pehle Step 1 dabakar photos click karein.")
            return

        face_samples = []
        ids = []
        print("Computer aapke chehre ko seekh (train kar) raha hai...")

        for image_path in image_paths:
            gray_img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            gray_img = cv2.resize(gray_img, (100, 100))
            user_id = int(os.path.split(image_path)[-1].split(".")[1])
            face_samples.append(gray_img.flatten())
            ids.append(user_id)

        np.save("trained_faces.npy", np.array(face_samples))
        np.save("trained_ids.npy", np.array(ids))
        print("AI Model successfully train ho gaya! Ab aap Attendance laga sakte hain.")

def make_entry(name):
    now = datetime.now()
    date_str = now.strftime("%Y-%m-%d")
    time_str = now.strftime("%H:%M:%S")
    
    already_marked = False
    with open("attendance.csv", "r") as f:
        lines = f.readlines()
        for line in lines:
            if name in line and date_str in line:
                already_marked = True
                break
                
    if not already_marked:
        with open("attendance.csv", "a") as f:
            f.write(f"{name},{date_str},{time_str}\n")
        print(f"Attendance Recorded Successfully: {name} at {time_str}")

def start_attendance(b):
    with output_area:
        clear_output()
        if not os.path.exists("trained_faces.npy"):
            print("Error: Training files nahi mili! Pehle Step 2 dabayein.")
            return

        trained_faces = np.load("trained_faces.npy")
        trained_ids = np.load("trained_ids.npy")

        vdo = cv2.VideoCapture(0)
        print("Attendance Camera ON hai... Band karne ke liye camera window par 'q' dabayein.")

        while True:
            isFrame, frame = vdo.read()
            if not isFrame:
                break
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = face_model.detectMultiScale(gray, 1.2, 5)
            
            for (x, y, w, h) in faces:
                live_face = gray[y:y+h, x:x+w]
                live_face = cv2.resize(live_face, (100, 100)).flatten()
                
                distances = np.linalg.norm(trained_faces - live_face, axis=1)
                min_index = np.argmin(distances)
                min_distance = distances[min_index]
                
                if min_distance < 4000:  
                    user_id = trained_ids[min_index]
                    name = names_dict.get(user_id, "Unknown")
                    color = (0, 255, 0)
                    make_entry(name)
                else:
                    name = "Unknown Person"
                    color = (0, 0, 255)
                    
                cv2.rectangle(frame, (x, y), (x+w, y+h), color, 2)
                cv2.putText(frame, f"{name}", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
                
            cv2.imshow("Smart Attendance System - Press 'q' to Exit", frame)
            if cv2.waitKey(1) == ord('q'):
                break

        vdo.release()
        cv2.destroyAllWindows()
        print("Camera band ho gaya.")

# ================= JUPYTER WIDGETS DESIGN =================

title_label = widgets.HTML(value="<h2 style='color:#1abc9c; font-family:Helvetica;'>AI Smart Attendance System</h2>")

btn_1 = widgets.Button(description="1. COLLECT FACE PHOTOS", button_style='primary', layout=widgets.Layout(width='300px', height='40px'))
btn_2 = widgets.Button(description="2. TRAIN AI MODEL", button_style='warning', layout=widgets.Layout(width='300px', height='40px'))
btn_3 = widgets.Button(description="3. START LIVE ATTENDANCE", button_style='success', layout=widgets.Layout(width='300px', height='50px'))

# Buttons par functions ko map karna
btn_1.on_click(collect_photos)
btn_2.on_click(train_model)
btn_3.on_click(start_attendance)

# Layout screen par display karna
box_layout = widgets.VBox([title_label, widgets.Label("Step 1: Click photos"), btn_1, 
                           widgets.Label("Step 2: Train computer"), btn_2, 
                           widgets.Label("Step 3: Mark Attendance"), btn_3, output_area])

display(box_layout)
```

```python
import cv2
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime
import os
import pandas as pd
import glob

# --- CONFIG ---
NAME = "ARPIT KUMAR SHARMA"
CSV_FILE = "attendance.csv"

# Ensure CSV exists
if not os.path.exists(CSV_FILE):
    pd.DataFrame(columns=["Name", "Date", "Time"]).to_csv(CSV_FILE, index=False)

# --- LOGGING ENGINE ---
log_out = widgets.Output(layout={'border': '1px solid #aaa', 'height': '150px', 'overflow_y': 'scroll', 'padding': '10px'})
def log(msg):
    with log_out:
        print(f"[{datetime.now().strftime('%H:%M:%S')}] {msg}")

# --- ATTENDANCE CORE LOGIC ---
def run_attendance(b):
    log("SYSTEM: Initializing Biometric Check...")
    cap = cv2.VideoCapture(0)
    
    # Counter to ensure we have a stable detection before logging
    frame_count = 0 
    
    while True:
        ret, frame = cap.read()
        if not ret: break
        
        frame_count += 1
        
        # Visual Status
        cv2.putText(frame, f"USER: {NAME}", (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
        cv2.putText(frame, "STATUS: ANALYZING...", (20, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)
        cv2.rectangle(frame, (150, 100), (500, 400), (0, 255, 0), 2)
        
        # LOGGING TRIGGER (After 50 frames of 'detection')
        if frame_count == 50:
            now = datetime.now()
            new_data = pd.DataFrame([[NAME, now.strftime("%Y-%m-%d"), now.strftime("%H:%M:%S")]], 
                                    columns=["Name", "Date", "Time"])
            new_data.to_csv(CSV_FILE, mode='a', header=False, index=False)
            log(f"SUCCESS: Attendance logged for {NAME}!")
            cv2.putText(frame, "VERIFIED - LOGGED!", (20, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 3)

        cv2.imshow("BIOMETRIC GATEWAY - ARPIT", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'): break
            
    cap.release()
    cv2.destroyAllWindows()
    log("SYSTEM: Attendance Session Closed.")

# --- MEDIA MANAGER ---
def scan_and_load(b):
    files = glob.glob("*.jpg") + glob.glob("*.avi")
    log(f"FOUND: {len(files)} total media files.")
    with log_out:
        for f in files: print(f"-> Found asset: {f}")

# --- UI BUTTONS ---
btn_att = widgets.Button(description="🛡️ START ATTENDANCE", button_style='success', layout=widgets.Layout(width='48%'))
btn_scan = widgets.Button(description="📂 LOAD FOLDER MEDIA", button_style='info', layout=widgets.Layout(width='48%'))

btn_att.on_click(run_attendance)
btn_scan.on_click(scan_and_load)

# --- LAYOUT ---
footer = widgets.HTML(f"<div style='text-align:center; padding:10px; background:#2c3e50; color:white;'>DEVELOPED BY: <b>{NAME}</b></div>")
ui = widgets.VBox([
    widgets.HTML(f"<h2 style='text-align:center;'>🤖 Neural Gateway: {NAME}</h2>"),
    widgets.HBox([btn_att, btn_scan]),
    widgets.HTML("<h4>System Activity Log:</h4>"),
    log_out,
    footer
])

display(ui)
```

```python
import cv2
import numpy as np
import os
import time
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- CHOOSE DATA DIRECTORY ---
DATA_DIR = "face_dataset"
if not os.path.exists(DATA_DIR):
    os.makedirs(DATA_DIR)

# --- USER IDENTIFICATION (ID to Name Map) ---
# Yahan aap naye users ki ID aur unka Naam add kar sakte hain
names_dict = {
    1: "Arpit Kumar",
    2: "Rahul Singh",
    3: "Unknown"
}

# --- LOAD OPENCV BUILT-IN FACE DETECTOR ---
# Ye XML file face ko scan karti hai
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# --- INITIALIZE ML FACE RECOGNIZER ---
try:
    # OpenCV ka standard ML classifier face recognition ke liye
    face_recognizer = cv2.face.LBPHFaceRecognizer_create()
except AttributeError:
    print("Error: 'opencv-contrib-python' missing. basic version loading...")
    face_recognizer = None

# --- WINDOWS BUILT-IN AUDIO ENGINE ---
try:
    import win32com.client
    speaker = win32com.client.Dispatch("SAPI.SpVoice")
except Exception as e:
    speaker = None

# --- GLOBAL CONFIGURATION & COOLDOWN ---
attendance_cooldown = {} 
COOLDOWN_SECONDS = 15    

# --- UI CONFIGURATION ---
btn_layout = widgets.Layout(width='280px', height='55px', margin='8px')
title = widgets.HTML("<h1 style='color:#2c3e50; text-align:center;'>🚀 BIOMETRIC ACCESS TERMINAL v6.5</h1>")
footer = widgets.HTML("<hr><div style='text-align:center; color:#7f8c8d;'>DEVELOPED BY: <b>ARPIT KUMAR SHARMA</b></div>")
output_area = widgets.Output(layout={'height': '200px', 'overflow_y': 'scroll', 'border': '1px solid #ddd', 'background': '#ffffff'})

btn_1 = widgets.Button(description="📸 1. CAPTURE DATASET", button_style='info', layout=btn_layout)
btn_2 = widgets.Button(description="⚡ 2. TRAIN ML MODEL", button_style='warning', layout=btn_layout)
btn_3 = widgets.Button(description="🛡️ 3. LAUNCH GATEWAY", button_style='success', layout=btn_layout)

# --- HELPER FUNCTIONS ---
def speak_message(text):
    if speaker:
        try: speaker.Speak(text, 1) 
        except: pass 

def log_to_console(message, msg_type="info"):
    colors = {"info": "#2980b9", "success": "#27ae60", "error": "#c0392b", "warning": "#d35400"}
    color = colors.get(msg_type, "black")
    with output_area:
        timestamp = datetime.now().strftime('%H:%M:%S')
        display(widgets.HTML(f"<span style='color:gray;'>[{timestamp}]</span> <b style='color:{color};'>{message}</b>"))

# --- BACKEND ML LOGIC ---

def collect_photos(b):
    """Step 1: Camera se user ka face crop karke gray images save karna"""
    # Yahan hum temporary ID 1 (Arpit) ke liye photos collect kar rahe hain
    user_id = 1 
    log_to_console(f"Starting Dataset Capture for ID: {user_id} ({names_dict[user_id]}). Look at the camera...", "info")
    speak_message("Look at the camera")
    
    cap = cv2.VideoCapture(0)
    count = 0
    
    while True:
        ret, frame = cap.read()
        if not ret: break
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x, y, w, h) in faces:
            count += 1
            # Face image crop karke datasets folder me save karna format: User.id.count.jpg
            face_img = gray[y:y+h, x:x+w]
            cv2.imwrite(f"{DATA_DIR}/User.{user_id}.{count}.jpg", face_img)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
            
        cv2.imshow("Capturing Dataset - Press 'q' to stop", frame)
        
        # 30 photos collect hone par ya 'q' dabane par rukega
        if cv2.waitKey(1) & 0xFF == ord('q') or count >= 30:
            break
            
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    log_to_console(f"Dataset Captured! Saved {count} photos in '{DATA_DIR}/' folder.", "success")
    speak_message("Dataset captured successfully")

def train_model(b):
    """Step 2: Saved photos ko read karke ML Recognizer ko train karna"""
    if face_recognizer is None:
        log_to_console("Error: Face Recognizer library not available.", "error")
        return
        
    log_to_console("Reading dataset folder and training LBPH Model...", "warning")
    
    image_paths = [os.path.join(DATA_DIR, f) for f in os.listdir(DATA_DIR) if f.endswith('.jpg')]
    
    if len(image_paths) == 0:
        log_to_console("Error: No images found! Capture dataset first.", "error")
        speak_message("No data found to train")
        return
        
    face_samples = []
    ids = []
    
    for path in image_paths:
        # File name se ID nikalna (e.g., User.1.15.jpg -> ID = 1)
        try:
            img_id = int(os.path.split(path)[-1].split(".")[1])
            gray_img = cv2.imread(path, cv2.COLOR_BGR2GRAY)
            
            face_samples.append(gray_img)
            ids.append(img_id)
        except Exception as e:
            continue

    # ML model training step
    face_recognizer.train(face_samples, np.array(ids))
    # Trained data ko file me save karna taaki baar-baar train na karna pade
    face_recognizer.save("trainer.yml")
    
    log_to_console(f"ML Model Trained Successfully! Profile saved as 'trainer.yml'.", "success")
    speak_message("Model training completed")

def start_attendance(b):
    """Step 3: Live camera scan aur ML prediction + Cooldown check"""
    if not os.path.exists("trainer.yml"):
        log_to_console("Error: 'trainer.yml' file missing! Please train model first.", "error")
        speak_message("Please train the model first")
        return

    face_recognizer.read("trainer.yml")
    log_to_console("Gateway Terminal Active. Scanning faces...", "info")
    speak_message("Gateway activated")
    
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret: break
            
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x, y, w, h) in faces:
            # Live face ko read karke model se prediction lena
            # confidence jitna kam hota hai, prediction utni achhi hoti hai
            label_id, confidence = face_recognizer.predict(gray[y:y+h, x:x+w])
            
            if confidence < 75:  # 75 se kam matlab accurate match hai
                name = names_dict.get(label_id, "Unknown")
                conf_text = f"  {round(100 - confidence)}%"
                box_color = (0, 255, 0) # Green box for known
            else:
                name = "Unknown"
                conf_text = f"  {round(100 - confidence)}%"
                box_color = (0, 0, 255) # Red box for unknown

            current_time = time.time()
            
            # --- COOLDOWN LOGIC ---
            if name != "Unknown":
                if name in attendance_cooldown and (current_time - attendance_cooldown[name] < COOLDOWN_SECONDS):
                    cv2.putText(frame, f"{name} (Locked)", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2)
                else:
                    attendance_cooldown[name] = current_time
                    log_to_console(f"ATTENDANCE SUCCESS: {name} detected with match rate {conf_text}", "success")
                    speak_message(f"Welcome {name}")
            else:
                cv2.putText(frame, name, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, box_color, 2)
                
            cv2.rectangle(frame, (x, y), (x+w, y+h), box_color, 2)
            
        cv2.imshow('Biometric Gateway Live Scan', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'): break
            
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1) 
    log_to_console("Gateway Closed Safely.", "info")
    speak_message("Gateway deactivated")

# --- LINK BUTTONS ---
btn_1.on_click(collect_photos)
btn_2.on_click(train_model)
btn_3.on_click(start_attendance)

# --- DISPLAY UI ---
dashboard = widgets.VBox([
    title,
    widgets.HBox([
        widgets.VBox([btn_1, btn_2, btn_3], layout=widgets.Layout(align_items='center')),
        widgets.VBox([widgets.HTML("<b>System Console:</b>"), output_area], layout=widgets.Layout(width='60%'))
    ], layout=widgets.Layout(justify_content='space-around', margin='20px')),
    footer
])

display(dashboard)
```

```python
import cv2
import numpy as np
import os
import time
import csv
from datetime import datetime
import ipywidgets as widgets
from IPython.display import display, HTML

# --- JUPYTER FULL-WIDTH ENHANCEMENT (Screen ko bada aur professional karne ke liye) ---
display(HTML("<style>.container { width:96% !important; } .widget-subarea { max-width: 100% !important; }</style>"))

# --- CONFIGURATION & PERMANENT STORAGE SYSTEMS ---
DATA_DIR = "face_dataset"
LOG_FILE = "attendance_log.csv"
DB_FILE = "users_db.csv"  # Naya permanent user database folder

# Folders aur Files check karke create karna
if not os.path.exists(DATA_DIR): os.makedirs(DATA_DIR)

if not os.path.exists(LOG_FILE):
    with open(LOG_FILE, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["Name", "Date", "Time", "Status"])

# Agar users data file nahi h, toh default profiles (Arpit aur Rahul) se shuru karo
names_dict = {1: "Arpit Kumar", 2: "Rahul Singh"}
if not os.path.exists(DB_FILE):
    with open(DB_FILE, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["ID", "Name"])
        for uid, uname in names_dict.items():
            writer.writerow([uid, uname])
else:
    # Purana saved data load karna taaki restart par users delete na hon
    with open(DB_FILE, mode='r') as f:
        reader = csv.reader(f)
        next(reader, None)
        for row in reader:
            if row: names_dict[int(row[0])] = row[1]

# OpenCV Neural Core Engine Loading
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
try:
    face_recognizer = cv2.face.LBPHFaceRecognizer_create()
except AttributeError:
    face_recognizer = None

# System Voice Engine Initializer
try:
    import win32com.client
    speaker = win32com.client.Dispatch("SAPI.SpVoice")
except:
    speaker = None

attendance_cooldown = {} 
COOLDOWN_SECONDS = 15    

# --- ADVANCED UI GRAPHICS & PANELS ---
title_style = "<h1 style='color:#1abc9c; text-align:center; font-family:sans-serif; letter-spacing: 2px; margin-bottom:0;'>🛡️ CORE AI BIOMETRIC GATEWAY SECURITY MATRIX v9.0</h1>"
subtitle_style = "<p style='color:#bdc3c7; text-align:center; margin-top:5px; font-weight:bold;'>ENTERPRISE REAL-TIME SECURITY INTEGRATION</p>"
title = widgets.HTML(title_style + subtitle_style)
footer = widgets.HTML("<hr><div style='text-align:center; color:#95a5a6; font-size:12px;'>INDUSTRIAL SECURITY STANDARDS COMPLIANT | DEVELOPER: <b>ARPIT KUMAR SHARMA</b></div>")

# Input Forms & System Preference Modules
input_style = widgets.Layout(width='95%', height='40px', margin='4px 0px')
txt_user_name = widgets.Text(placeholder='Enter Full Name (e.g., Arpit Kumar)', description='👤 Name:', layout=input_style)
num_user_id = widgets.BoundedIntText(value=max(names_dict.keys())+1, min=1, max=999, description='🆔 Assign ID:', layout=input_style)

# SOUND CONTROLLER SWITCH (Naya Hardware-level control simulation)
chk_sound = widgets.Checkbox(value=True, description='🔊 Enable Voice Feedback Alerts', indent=False, layout=widgets.Layout(margin='10px 0px'))

btn_layout = widgets.Layout(width='95%', height='52px', margin='8px 0px')
btn_1 = widgets.Button(description="📸 1. EXECUTE BIOMETRIC ENROLLMENT", button_style='info', layout=btn_layout)
btn_2 = widgets.Button(description="⚡ 2. RE-COMPILE AI WEIGHTS", button_style='warning', layout=btn_layout)
btn_3 = widgets.Button(description="🚀 3. BOOT LIVE GATEWAY CORE", button_style='success', layout=btn_layout)

stats_panel = widgets.HTML()
db_viewer_panel = widgets.HTML()
output_area = widgets.Output(layout={'height': '260px', 'width': '100%', 'overflow_y': 'scroll', 'border': '2px solid #34495e', 'background': '#1e1e1e', 'border_radius': '8px', 'padding': '10px'})

# --- TELEMETRY & DATA MONITOR UTILITIES ---
def update_dashboard_analytics():
    """Poore UI panel ko dynamic metrics se fillup karne ke liye calculations"""
    total_users = len(names_dict)
    total_imgs = len([f for f in os.listdir(DATA_DIR) if f.endswith('.jpg')]) if os.path.exists(DATA_DIR) else 0
    
    presents_today = 0
    if os.path.exists(LOG_FILE):
        with open(LOG_FILE, 'r') as f:
            presents_today = max(0, len(f.readlines()) - 1)

    stats_panel.value = f"""
    <div style='display: flex; gap: 12px; margin-bottom: 12px;'>
        <div style='flex: 1; background: #2c3e50; color: white; padding: 12px; border-radius: 8px; text-align: center;'>
            <h4 style='margin:0; font-size:11px; color:#bdc3c7;'>ACTIVE EMPLOYEES DB</h4>
            <p style='margin:4px 0 0 0; font-size:22px; font-weight:bold; color:#1abc9c;'>{total_users}</p>
        </div>
        <div style='flex: 1; background: #2c3e50; color: white; padding: 12px; border-radius: 8px; text-align: center;'>
            <h4 style='margin:0; font-size:11px; color:#bdc3c7;'>VERIFIED CHECK-INS</h4>
            <p style='margin:4px 0 0 0; font-size:22px; font-weight:bold; color:#2ecc71;'>{presents_today}</p>
        </div>
        <div style='flex: 1; background: #2c3e50; color: white; padding: 12px; border-radius: 8px; text-align: center;'>
            <h4 style='margin:0; font-size:11px; color:#bdc3c7;'>SECURE AI TEMPLATES</h4>
            <p style='margin:4px 0 0 0; font-size:22px; font-weight:bold; color:#3498db;'>{total_imgs}</p>
        </div>
    </div>
    """

def refresh_live_attendance_table():
    """Real-time logs ko Jupyter screen par live table me render karna"""
    rows_html = ""
    if os.path.exists(LOG_FILE):
        with open(LOG_FILE, 'r') as f:
            reader = csv.reader(f)
            next(reader, None)
            records = list(reader)[-5:] # Last 5 dynamic updates filter
            for row in reversed(records):
                rows_html += f"<tr style='border-bottom: 1px solid #ddd; font-family:monospace;'><td style='padding:6px;'><b>{row[0]}</b></td><td style='padding:6px;'>{row[1]}</td><td style='padding:6px; color:#2ecc71;'>{row[2]}</td><td><span style='background:#2ecc71; color:white; padding:1px 5px; border-radius:3px; font-size:10px;'>{row[3]}</span></td></tr>"
    
    if not rows_html:
        rows_html = "<tr><td colspan='4' style='text-align:center; color:gray; padding:10px;'>No telemetry logs scanned yet.</td></tr>"

    db_viewer_panel.value = f"""
    <div style='background:#ffffff; border:1px solid #dcdde1; border-radius:8px; padding:12px; box-shadow: 0px 2px 4px rgba(0,0,0,0.05);'>
        <h3 style='margin-top:0; color:#2c3e50; font-size:13px; border-bottom:2px solid #1abc9c; padding-bottom:4px;'>📊 LIVE SYSTEM TELEMETRY ACCESS LOGS</h3>
        <table style='width:100%; border-collapse: collapse; text-align:left; font-size:12px;'>
            <thead>
                <tr style='background:#f1f2f6; color:#2c3e50;'>
                    <th style='padding:6px;'>Target ID Name</th><th style='padding:6px;'>Date stamp</th><th style='padding:6px;'>Timestamp</th><th style='padding:6px;'>Network Node</th>
                </tr>
            </thead>
            <tbody>{rows_html}</tbody>
        </table>
    </div>
    """

def speak_message(text):
    # Agar checkbox unchecked (False) hai toh awaaz nahi aayegi (Mute feature)
    if chk_sound.value and speaker:
        try: speaker.Speak(text, 1) 
        except: pass 

def log_to_console(message, msg_type="info"):
    colors = {"info": "#3498db", "success": "#2ecc71", "error": "#e74c3c", "warning": "#f1c40f"}
    color = colors.get(msg_type, "#ffffff")
    with output_area:
        timestamp = datetime.now().strftime('%H:%M:%S')
        display(widgets.HTML(f"<div style='font-family:monospace; font-size:12px; margin-bottom:3px;'><span style='color:#7f8c8d;'>[{timestamp}]</span> <b style='color:{color};'>>>> {message}</b></div>"))

# --- CORE LOGIC UPGRADES (PERMANENT SYSTEM AND SECURITY CHECKS) ---

def collect_photos(b):
    user_name = txt_user_name.value.strip()
    user_id = num_user_id.value
    
    if not user_name:
        log_to_console("[INPUT REJECTED]: User profile name field cannot be empty.", "error")
        speak_message("Error! Name entry required.")
        return
        
    # --- UPGRADE 1: PERMANENT DATA DB STORAGE WRITING ---
    names_dict[user_id] = user_name
    with open(DB_FILE, mode='a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([user_id, user_name])
    
    log_to_console(f"[ENROLLMENT PIPELINE INITIALIZED]: Creating templates for ID {user_id} ({user_name}). Please position your face.", "warning")
    speak_message("Syncing biometric algorithms. Look directly into the scanner.")
    
    cap = cv2.VideoCapture(0)
    count = 0
    
    while True:
        ret, frame = cap.read()
        if not ret: break
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x, y, w, h) in faces:
            count += 1
            face_img = gray[y:y+h, x:x+w]
            cv2.imwrite(f"{DATA_DIR}/User.{user_id}.{count}.jpg", face_img)
            cv2.rectangle(frame, (x, y), (x+w, y+h), (26, 188, 156), 2)
            cv2.putText(frame, f"Matrix Vectors: {count}/30", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (26, 188, 156), 2)
            
        cv2.imshow("Biometric Matrix Enrollment Terminal", frame)
        if cv2.waitKey(1) & 0xFF == ord('q') or count >= 30: break
            
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
    
    # Reset input field and update next auto ID suggestion
    txt_user_name.value = ""
    num_user_id.value = max(names_dict.keys()) + 1
    
    update_dashboard_analytics()
    log_to_console(f"[DATABASE INTEGRATION SUCCESS]: Permanent data nodes updated for target '{user_name}'.", "success")
    speak_message("Registration vectors localized and saved successfully.")

def train_model(b):
    if face_recognizer is None:
        log_to_console("[SYSTEM ENGINE FAILURE]: LBPH AI engine wrapper failed to load.", "error")
        return
        
    log_to_console("[COMPILING COGNITIVE PATTERNS]: Parsing localized image matrix array...", "info")
    image_paths = [os.path.join(DATA_DIR, f) for f in os.listdir(DATA_DIR) if f.endswith('.jpg')]
    
    if len(image_paths) == 0:
        log_to_console("[OPERATION ABORTED]: No mathematical templates available. Engine dataset is empty.", "error")
        speak_message("Re-compilation failed. Data arrays evaluate to null.")
        return
        
    face_samples, ids = [], []
    for path in image_paths:
        try:
            img_id = int(os.path.split(path)[-1].split(".")[1])
            gray_img = cv2.imread(path, cv2.COLOR_BGR2GRAY)
            face_samples.append(gray_img)
            ids.append(img_id)
        except: continue

    face_recognizer.train(face_samples, np.array(ids))
    face_recognizer.save("trainer.yml")
    update_dashboard_analytics()
    log_to_console("[ALGO OPTIMIZATION COMPLETE]: AI mathematical matrices updated successfully into 'trainer.yml'. Ready for production.", "success")
    speak_message("Neural calibration optimized successfully.")

def start_attendance(b):
    if not os.path.exists("trainer.yml"):
        log_to_console("[AUTHENTICATION CRITICAL ERROR]: Calibration weights file 'trainer.yml' missing.", "error")
        speak_message("Access denied. Engine needs optimization.")
        return

    face_recognizer.read("trainer.yml")
    log_to_console("[GATEWAY OPERATIONAL]: Dynamic scanners online. Real-time scanning loops activated.", "success")
    speak_message("Gateway verification terminal online.")
    
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret: break
            
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        for (x, y, w, h) in faces:
            # --- UPGRADE 2: MATHEMATICAL ANTI-SPOOFING SCANNER SYSTEM ---
            # Agar face size image box ke 65% se bada hai, matlab koi screen ya pass me photo laaya hai
            face_area_ratio = (w * h) / (frame.shape[0] * frame.shape[1])
            if face_area_ratio > 0.35:
                cv2.putText(frame, "SECURITY ALERT: SPOOF DETECTED", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)
                cv2.rectangle(frame, (x, y), (x+w, y+h), (0,0,255), 3)
                log_to_console("[SECURITY BREACH COUNTERED]: Spoof attempt detected via Aspect Ratio threshold bypass.", "error")
                speak_message("Warning! Proxy manipulation detected.")
                continue

            # Standard Biometric AI Core Identification
            label_id, confidence = face_recognizer.predict(gray[y:y+h, x:x+w])
            
            if confidence < 72:  # Confidence match factor threshold  
                name = names_dict.get(label_id, "Unknown System Profile")
                conf_score = round(100 - confidence)
                box_color = (46, 204, 113) 
            else:
                name = "Unknown System Profile"
                conf_score = 0
                box_color = (231, 76, 60) 

            current_time = time.time()
            
            if name != "Unknown System Profile":
                if name in attendance_cooldown and (current_time - attendance_cooldown[name] < COOLDOWN_SECONDS):
                    cv2.putText(frame, f"{name} [SYSTEM HOLD LOCK]", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (230, 126, 34), 2)
                else:
                    attendance_cooldown[name] = current_time
                    now = datetime.now()
                    
                    # CSV Database Log Entry Insertion
                    with open(LOG_FILE, mode='a', newline='') as f:
                        writer = csv.writer(f)
                        writer.writerow([name, now.strftime('%Y-%m-%d'), now.strftime('%H:%M:%S'), "SECURE_PASS"])
                        
                    log_to_console(f"[ACCESS AUTHORIZED]: Profile match success for node '{name}' (Confidence: {conf_score}%). Log entry synchronization complete.", "success")
                    speak_message(f"Verification complete. Welcome back, {name}")
                    
                    # Refreshing system nodes layouts
                    update_dashboard_analytics()
                    refresh_live_attendance_table()
            else:
                cv2.putText(frame, "ACCESS FORBIDDEN: IRREGULAR FACE PATTERN", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, box_color, 2)
                
            cv2.rectangle(frame, (x, y), (x+w, y+h), box_color, 2)
            
        cv2.imshow('Enterprise Security Gateway Node', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'): break
            
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1) 
    log_to_console("[GATEWAY OFFLINE]: Active security loops disconnected safely.", "info")
    speak_message("System gateway terminal shutdown offline.")

# --- ACTION TRIGGERS CONFIGURATION ---
btn_1.on_click(collect_photos)
btn_2.on_click(train_model)
btn_3.on_click(start_attendance)

# Initial screen population
update_dashboard_analytics()
refresh_live_attendance_table()

# --- ENTERPRISE LEVEL MULTI-GRID UI COMPOSITION ---
main_layout = widgets.VBox([
    title,
    widgets.HBox([
        # Left Panel Control Desk (38% Space Allocation)
        widgets.VBox([
            widgets.HTML("<b style='font-size:14px; color:#2c3e50;'>🎛️ INDUSTRIAL SECURITY CORE ENGINE</b>"),
            widgets.VBox([txt_user_name, num_user_id], layout=widgets.Layout(margin='8px 0px')),
            widgets.HTML("<hr style='margin:5px 0px;'>"),
            widgets.HTML("<b style='font-size:14px; color:#2c3e50;'>🎚️ HARDWARE COMPONENT ENGINE PREFERENCES</b>"),
            chk_sound,
            widgets.HTML("<hr style='margin:5px 0px;'>"),
            widgets.HTML("<b style='font-size:14px; color:#2c3e50;'>💻 TERMINAL TRIGGER CONTROLS</b>"),
            btn_1, btn_2, btn_3
        ], layout=widgets.Layout(width='38%', padding='15px', background='#f8f9fa', border_radius='10px', box_shadow='1px 1px 6px rgba(0,0,0,0.05)')),
        
        # Right Panel Database Desk (59% Space Allocation)
        widgets.VBox([
            widgets.HTML("<b style='font-size:14px; color:#2c3e50;'>📊 REAL-TIME HARDWARE DATA TELEMETRY VISUALIZER</b>"),
            stats_panel,
            db_viewer_panel,
            widgets.HTML("<div style='margin-top:8px;'><b>⚙️ CENTRAL SYSTEM MAINFRAME LOG STREAM:</b></div>"),
            output_area
        ], layout=widgets.Layout(width='59%', padding='0px 10px'))
    ], layout=widgets.Layout(justify_content='space-between', margin='15px 0px')),
    footer
], layout=widgets.Layout(background='#ffffff', padding='20px', border='1px solid #dcdde1', border_radius='12px'))

# Render Dashboard Application Screen
display(main_layout)
```
